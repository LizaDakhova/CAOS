```python
import math
```

```python
def f(x):
    return math.sin(x[0] / 5) * math.exp(x[0] / 10) +  5 * math.exp(-x[0] / 2)
```

```python
%matplotlib inline
```

```python
from matplotlib import pylab as plt
```

```python
import numpy as np
```

```python
x = np.arange(1, 15, 0.01)
yf = [f([xi]) for xi in x]
plt.plot(x, yf)
plt.show()
```

```python
print f([1]), f([15])
```

### Сиcтема линейных уравнений, приюлижающая функцию
Составляем систему по точкам 1 и 15.

```python
a1 = np.array([[1, 1], [15, 1]])
b1 = np.array([f([1]), f([15])])
z1 = np.linalg.solve(a1, b1)
```

```python
print z1
```

```python
y1 = x * z1[0] + z1[1]
```

```python
plt.plot(x, yf, x, y1)
plt.show()
```

```python
a2 = np.array([[1, 1, 1], [64, 8, 1], [225, 15, 1]])
b2 = np.array([f([1]), f([8]), f([15])])
z2 = np.linalg.solve(a2, b2)
```

```python
print z2
```

```python
y2 = x ** 2 * z2[0] + x * z2[1] + z2[2]
```

```python
plt.plot(x, yf, x, y2)
plt.show()
```

```python
a3 = np.array([[1, 1, 1, 1], [64, 16, 4, 1], [1000, 100, 10, 1], [225 * 15, 225, 15, 1]])
b3 = np.array([f([1]), f([4]), f([10]), f([15])])
z3 = np.linalg.solve(a3, b3)
```

```python
print z3
```

```python
y3 = x ** 3 * z3[0] + x ** 2 * z3[1] + x * z3[2] + z3[3]
```

```python
plt.plot(x, yf, x, y3)
plt.show()
```
