```python
import random as rand
import string
```

```python
k = 256
size = 10
a = []
for i in range(k):
    a.append(''.join([rand.choice(string.ascii_letters + string.digits) for _ in range(size)]))
    f = open(a[i], 'w')
    rand_size0 = rand.randint(1, 100)
    for j in range(rand_size0):
        rand_size1 = rand.randint(1, 30)
        s = ''.join([rand.choice(string.ascii_letters + string.digits) for _ in range(rand_size1)]) + '\n'
        f.write(s)
    f.close()
    
f = open("many_files", 'w')
for i in range(k):
    f.write(a[i] + '\n')
f.close()
```
